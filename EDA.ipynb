```python
from sklearn.metrics import matthews_corrcoef
import numpy as np

# find correlation between symptoms  
corr_matrix = training_data.drop(columns=["prognosis"])

n_sample, n_feature = corr_matrix.shape
rff_all = []
for i in range(n_feature):
    for j in range(i + 1, n_feature):
        # Ensure i and j are valid indices
        if not (0 <= i < corr_matrix.shape[1]) or not (0 <= j < corr_matrix.shape[1]):
            raise IndexError("i or j is out of bounds")
        coeff_f_f = abs(matthews_corrcoef(corr_matrix[:, i], corr_matrix[:, j]))
        print(coeff_f_f)
        rff_all.append(coeff_f_f)
rff = np.mean(rff_all)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=0)

X_train = training_data.drop(columns=["prognosis"])
y_train = training_data["prognosis"]

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

X_test = testing_data.drop(columns=["prognosis"])
y_test = testing_data["prognosis"]

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

conf_m = confusion_matrix(y_test, y_pred)
print(conf_m)
```

```python
import seaborn as sns

plt.figure(figsize=(10, 10))
sns.heatmap(conf_m, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("True")

plt.show()
```
